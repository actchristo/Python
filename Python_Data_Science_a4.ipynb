```python
import pandas as pd
import numpy as np
import io
import folium

from google.colab import files

import plotly.express as px
```

```python
# Faz o upload do arquivo do computador para a nuvem
file_uploaded = files.upload()
```

```python
# Carrega o arquivo da nuvem para a IDE do Google
df = pd.read_csv(io.BytesIO(file_uploaded['AB_NYC_2019.csv']))
```

```python
# Remove as linhas vazias
df1 = df.dropna()
```

```python
df1.head()
```

```python
# Remover imoveis com a coluna availability_365 igual a zero

df1 = df1.loc[df1['availability_365'] > 0, :]
```

```python
# Ex 1 - Valor aluguel mais caro por Bairro de imóveis disponíveis
# Resp. : Bronx U$ 800, Brooklyn U$ 8000, Manhattan U$ 9999, Queens U$ 2600 e Staten Island U$ 625.

cols = ['neighbourhood_group', 'price']

data_plot = df1.loc[:, cols].groupby('neighbourhood_group').max().reset_index()

px.bar (data_plot, x='neighbourhood_group', y='price')
```

```python
# EX 2 - Onde estão localizados esses imóveis?
cols = ['price', 'neighbourhood_group', 'latitude', 'longitude']

data_plot = df1.loc[:, cols].groupby('neighbourhood_group').max().reset_index()

# Comando para desenhar um mapa
map = folium.Map() 

# Colocar os pontos no mapa
for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'], location_info['longitude']], popup = location_info ['neighbourhood_group']).add_to(map)

map
```

```python
# EX 3 - Mostrar localização por tipo

cols = ['neighbourhood_group', 'room_type', 'latitude', 'longitude']

data_plot = df1.loc[:, cols].sample(100) # 100 pontos aleatórios da tabela

# Criar uma nova coluna
data_plot.loc[:, 'color'] = 'NA'
data_plot.loc[data_plot['room_type'] == 'Private room','color'] = 'green'
data_plot.loc[data_plot['room_type'] == 'Entire home/apt','color'] = 'red'
data_plot.loc[data_plot['room_type'] == 'Shared room','color'] = 'purple'

# Comando para desenhar um mapa
#map = folium.Map() 

map = folium.Map(location=[40.64749, -73.97237], zoom_start = 11)

# Colocar os pontos no mapa
for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'], location_info['longitude']], 
                popup = location_info ['neighbourhood_group'], 
                icon=folium.Icon(color = location_info['color'])).add_to(map)

map
```
